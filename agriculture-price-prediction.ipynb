```python
cols=['Crop', 'State']
```

```python
for i in cols:
    print(i +': ')
    print(df[i].unique())
```

```python
df1=df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in cols:
    df1[i]=le.fit_transform(df1[i])
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df1.iloc[:,0:5]
```

```python
y=df1.iloc[:,-1:]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3,shuffle=True,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sr=StandardScaler()
```

```python
x_test.head()
```

```python
sc_cols=['Cost of Cultivation (`/Hectare) A2+FL',
       'Cost of Cultivation (`/Hectare) C2',
       'Cost of Production (`/Quintal) C2']
for j in sc_cols:
    x_train[j]=sr.fit_transform(x_train[[j]])
    x_test[j]=sr.transform(x_test[[j]])
```

```python
x_test
```

```python
y_train.values
```

```python
from sklearn.svm import SVR
svr=SVR()
model=svr.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(n_estimators=90,max_depth=10, random_state=0)
rfg=regr.fit(x_train, y_train)
```

```python
x_test
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred=rfg.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```
